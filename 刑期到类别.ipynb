```python
import pickle
with open("imp_class.dict",'rb') as f:
    imp_class=pickle.load(f)
```

```python
def imp2class(x,imp_dict):
    
    for key in   imp_dict.keys():
        
        s=key[1:-1].split(",")
        start=int(s[0])
        end=int(s[1])
        #print(start,end,imp_dict[key])
        
        if x>start-1 and x<end+1 :
            return int(imp_dict[key])
    

imp2class(1,imp_class)
```

```python
data.loc[data.death_penalty==True,'imprisonment']=-2
data.loc[data.life_imprisonment==True,'imprisonment']=-1
```

```python
data['imp_label']=data['imprisonment'].map(lambda x:imp2class(x,imp_class)).astype(int)
```

```python

data.loc[data.death_penalty==True,'imp_label']=0
data.loc[data.life_imprisonment==True,'imp_label']=1
```
